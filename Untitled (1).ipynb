```r
library(tidyverse)
install.packages('ggcorrplot')
library(ggcorrplot)  #Allows to render correlation plot

```

```r
rice_data <- read.csv("Rice_Cammeo_Osmancik.arff", header=FALSE, comment.char = "@")


clean_data <- rice_data[-c(0, 1, 2, 3, 4), ]

clean_data_2 <- type.convert(clean_data, as.is= TRUE)

head(clean_data_2)

binary_data<- clean_data_2 |>
     mutate(V8 = ifelse(V8 == "Cammeo",1,0))|>
    cor(use="complete.obs") 

head(binary_data)
tail(binary_data)

ggcorrplot(binary_data, 
           hc.order = TRUE, 
           type = "lower",
           lab = TRUE)
```
